```python
from nltk.corpus import wordnet

def wordnet_pos(tag):
    """Map a Brown POS tag to a WordNet POS tag."""
    
    table = {"N": wordnet.NOUN, "V": wordnet.VERB, "R": wordnet.ADV, "J": wordnet.ADJ}
    
    # Default to a noun.
    return table.get(tag[0], wordnet.NOUN)
```

```python
# Standardizing Text: Remove prefixes and suffixes from words through lemmatization.
tags = nltk.pos_tag(words)
tags
```

```python
[lemmatizer.lemmatize(w, wordnet_pos(t)) for (w, t) in tags]
```

```python
# Calculate the tf-idf (use sublinear_tf = True).

from sklearn.feature_extraction.text import TfidfVectorizer

vec = TfidfVectorizer(tokenizer = nltk.word_tokenize, sublinear_tf = True)
tfidf = vec.fit_transform(comment)
```
